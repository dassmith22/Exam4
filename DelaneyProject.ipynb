```python
'''
Imports required Libraries
'''
import csv
import pandas as pd
import re

'''
Opens file and puts each string into a list
'''
lists = []
with open("C:/Users/student/Documents/Test.csv", 'r') as file:
    reader = csv.reader(file, delimiter = '\t', quotechar=None)
    for val in reader:
        lists.append(str(val))
        
        

 
'''
using filter() to remove special characters
'''
special_char = "[]"
out_list = [''.join(filter(lambda i: i not in special_char, string)) for string in lists]  




'''
Remove character from Strings list
using loop + replace() + enumerate()
'''

char = "'"
for idx, ele in enumerate(out_list):
        out_list[idx] = ele.replace(char, '')
        
'''
Remove CSV column name, set size variable and set stringLength variable
'''

out_list.pop(0)

size = len(out_list)

stringLength = len(out_list[1])

```

```python
'''
This is the calculation process of possible Kmers
'''
def possK (k):
    option1 = pow(4,k)
    option2 = (stringLength - (k-1))
    if option1 > option2:
        return(option2)
    else:
        return(option1)
```
